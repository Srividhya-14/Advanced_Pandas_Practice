Ecommerce Sales Analysis

This notebook analyzes synthetic e-commerce sales data to understand performance across regions, product categories, and salespersons.


```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

## Task 1: Basic Grouping and Aggregations

This section analyzes the dataset using simple groupby operations to summarize sales by region, category, and salesperson.

1) Calculates the total sales revenue generated in each region and sorts the results to identify the highest performing region.

```python
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
region_sales
```

2. Counts how many transactions occurred in each product category.

```python
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

category_transactions
```

3. Calculates the average sales value generated by each salesperson.

```python
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()

salesperson_avg
```

4. For each result, we have used reset_index() to convert the result into a clean DataFrame.

5. Sort the regional sales results in descending order using sort_values(ascending=False) to identify the top-performing region

```python
region_sales_sorted = region_sales.sort_values(by='sales_amount', ascending=False)

region_sales_sorted
```

## Task 2: Multi-Column Grouping and Multiple Aggregations

Perform advanced grouping analysis:

1. Group by both region AND category to calculate total sales for each combination using:

```python
sales_region_category_grouped = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

sales_region_category_grouped
```

2. For each salesperson, calculate three metrics simultaneously using the agg() method:
Total sales ('sum')
Average sales ('mean')
Number of transactions ('count')
Use this syntax:

```python
salesperson_metrics = df.groupby('salesperson')['sales_amount'].agg(['sum','mean','count']).reset_index()
salesperson_metrics
```

3. Sort the salesperson results by total sales in descending order to identify the top performer

```python
salesperson_metrics_sorted = salesperson_metrics.sort_values(by='sum', ascending=False)
salesperson_metrics_sorted.head(1)
```

4. Use .idxmax() on the grouped category sales to find which category has the maximum total revenue

```python
category_sales = df.groupby('category')['sales_amount'].sum()
top_category = category_sales.idxmax()
top_category
```

##Task 3: Custom Aggregation and Complete Sales Report

1. Define a custom aggregation function that calculates the sales range (max - min) for each group:



```python
def sales_range(x):
    return x.max() - x.min()
```

2. Apply this custom function along with standard aggregations to analyze sales by region:



```python
analysis_by_region = df.groupby('region')['sales_amount'].agg(['sum','mean','min','max',sales_range]).reset_index()
analysis_by_region
```

3. Create a final summary report that shows for each region:

Total number of transactions (using customer_id with count)
Total sales amount
Average transaction value
Use dictionary syntax in agg():

```python
summary = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()
summary
```

4. Explain the multi-level column structure that results from the dictionary-based aggregation and how it differs from single aggregations

The dictionary-based aggregation creates a **multi-level column index** because multiple metrics are calculated for the same column.

For example:
- sales_amount → sum
- sales_amount → mean

This produces hierarchical column names such as:

sales_amount | sum  
sales_amount | mean  

This differs from single aggregations because a simple groupby with one function produces a single flat column.

region | sales_amount
